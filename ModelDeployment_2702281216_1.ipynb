```python
df = df.dropna()
```

```python
df = df.drop('Booking_ID', axis = 1)
```

```python
from sklearn import preprocessing

label_encoding = preprocessing.LabelEncoder()
df['booking_status'] = label_encoding.fit_transform(df['booking_status'])
df
```

```python
from sklearn.model_selection import train_test_split
x = df.drop('booking_status',axis=1)
y = df['booking_status']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Inisialisasi OneHotEncoder
enc = OneHotEncoder(sparse_output=False)

# Encoding untuk kolom kategorikal sekaligus
encoded_cols_train = enc.fit_transform(x_train[['type_of_meal_plan', 'room_type_reserved', 'market_segment_type']])
encoded_cols_test = enc.transform(x_test[['type_of_meal_plan', 'room_type_reserved', 'market_segment_type']])

# Konversi hasil encoding ke DataFrame
xtrain_prof_enc = pd.DataFrame(encoded_cols_train, columns=enc.get_feature_names_out(['type_of_meal_plan', 'room_type_reserved', 'market_segment_type']))
xtest_prof_enc = pd.DataFrame(encoded_cols_test, columns=enc.get_feature_names_out(['type_of_meal_plan', 'room_type_reserved', 'market_segment_type']))

# Reset index hanya sekali
x_train = x_train.reset_index(drop=True)
x_test = x_test.reset_index(drop=True)

# Gabungkan hasil encoding dengan data asli dan hapus kolom yang sudah di-encode
x_train = pd.concat([x_train, xtrain_prof_enc], axis=1).drop(['type_of_meal_plan', 'room_type_reserved', 'market_segment_type'], axis=1)
x_test = pd.concat([x_test, xtest_prof_enc], axis=1).drop(['type_of_meal_plan', 'room_type_reserved', 'market_segment_type'], axis=1)
```

```python
x_train['arrival_year'] = label_encoding.fit_transform(x_train['arrival_year'])
x_test['arrival_year'] = label_encoding.fit_transform(x_test['arrival_year'])
```

```python
from sklearn.preprocessing import RobustScaler

# Inisialisasi RobustScaler
scaler = RobustScaler()

# Scaling data
x_train_robust = scaler.fit_transform(x_train)
x_test_robust = scaler.transform(x_test)
```

```python
# Import library
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Training Model RandomForest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train_robust, y_train)

# Predicting dengan RandomForest
rf_pred = rf_model.predict(x_test_robust)

# Evaluasi Model RandomForest
rf_accuracy = accuracy_score(y_test, rf_pred)
print("RandomForest Accuracy: ", rf_accuracy)
print("RandomForest Classification Report:\n", classification_report(y_test, rf_pred))
```

```python
# Import library
import xgboost as xgb

# Training Model XGBoost
xgboost_model = xgb.XGBClassifier(n_estimators=100, random_state=42)
xgboost_model.fit(x_train_robust, y_train)

# Predicting dengan XGBoost
xgboost_pred = xgboost_model.predict(x_test_robust)

# Evaluasi Model XGBoost
xgboost_accuracy = accuracy_score(y_test, xgboost_pred)
print("XGBoost Accuracy: ", xgboost_accuracy)
print("XGBoost Classification Report:\n", classification_report(y_test, xgboost_pred))
```
